```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
m = plt.imread('IMG_0315.JPG')
plt.imshow(m)
plt.show()
```

```python
m2 = m[100:300, 200:400, :]
plt.imshow(m2)
plt.show()
```

```python
m3 = m[300:500, 400:600, :]
plt.imshow(m3)
plt.show()
```

```python
m4 = m2 / 2 + m3 / 2
plt.imshow(m4)
plt.show()
```

```python
plt.imshow(m4, cmap='gray')
plt.show()
```

```python
def scale(k, image):
    return image * k
plt.imshow(scale(0.5, m))
plt.show()
```
